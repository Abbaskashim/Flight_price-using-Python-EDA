```python
a= df.groupby(['Airline'], as_index=False)['Price'].sum().sort_values(by='Price',ascending=False)
a
```

```python
plt.figure(figsize=(15, 10))
x=sns.barplot(x='Airline', y='Price' , data=a)
for bars in x.containers:
    x.bar_label(bars, fmt='%.2f')

plt.title('Total Price by Airline')
plt.xlabel('Airline')
plt.ylabel('Total Price')
plt.xticks(rotation=90)
plt.show()
```
